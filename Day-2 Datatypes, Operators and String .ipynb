```python
# output formatting
a,b=10,20

print("The value of a is {} and b is {}".format(a,b))

print("Total students : %3d, Boys : %2d" % (240,120))

# print octal
print("%7.3o" % 25)

print("Geeks : %2d, Portal :%5.2f" % (1, 05.333)) 

# print exponential
print("%10.3E" % (356.08977))



# combining positional and keyword arguments
print('Number one portal is {0}, {1}, and {other}.'
     .format('Geeks', 'For', other ='Geeks'))
 
# using format() method with number 
print("Geeks :{0:2d}, Portal :{1:8.2f}".format(12, 00.546))

# Changing positional argument
print("Second argument: {1:3d}, first one: {0:7.2f}".format(47.42, 11))

print("Geeks : {a:5d}, Portal : {p:8.2f}".format(a=453, p=59.058))
```

```python
#String manupulation
s1 = "Python"
# 
print(s1[0]) # P
print(s1[-1]) #n

#len()

print(len(s1)) #6

list = [1,2,3,5]
print(len(list)) #4

# print(len(23)) # TypeError: object of type 'int' has no len()

#Type conversion

#1.Implicit

x = 10
print("x is of type:",type(x))

y = 10.6
print("y is of type:",type(y))

z = x + y
print(z)
print("z is of type:",type(z)) #z is of type: <class 'float'>


#Python Type conversion using  ord(), hex(), oct()


# initializing integer
s = '4'
 
# printing character converting to integer
c = ord(s) #the Unicode code from a given character
print ("After converting character to integer : ",end="")
print (c)

# printing integer converting to hexadecimal string
c = hex(56)
print ("After converting 56 to hexadecimal string : ",end="")
print (c)
 
# printing integer converting to octal string
c = oct(56)
print ("After converting 56 to octal string : ",end="")
print (c)


print("-----------------------------------------------")

# Python Type conversion using  tuple(), set(), list()
s = 'Python'
print("Tuple conversion :", tuple(s))
print("Set conversion :", set(s))
print("List conversion :", set(s))


print('---------------------------------------------')
# Type conversion using  dict(), complex(), str()

c = complex(1,2)
print("Complex num is: ",c) #(1+2j)


a = 7
print(type(a)) #<class 'int'>
c = str(a)
print(c)
print(type(c)) #<class 'str'>

tup = (('a',1),('b',2),('c',3))
print(tup)

c = dict(tup)
print("Dict is :",c) #{'a': 1, 'b': 2, 'c': 3}

print("---------------------------------------")
# f-string
weight=100
height=5.6

print(f"My weight is {weight} and height is {height}.")
```

```python
# Bill Calculator
print("Welcome to the tip calculator.")

totalBill = float(input("What was the total bill? $"))

totalPeople = int(input("How many people have to split the bill? "))

percentageTip = int(input("What percentage of tip would you like to give? 10,12 or 15? "))

perPersonPay = totalBill/totalPeople

eachPayWithTip = perPersonPay*(1+percentageTip*0.01)

print("Each person should pay $%.2f"%(eachPayWithTip))
```

```python
# Slice() in python

s = 'python'
s1 = slice(3)
s2 = slice(1, 5, 2)
s3 = slice(-1, -12, -2)

print(s[s1]) #pyt
print(s[s2]) #yh
print(s[s3]) 

print('------------------------------')
# [::] method

print(s[:3])
print(s[1:5:2])
print(s[-1:-12:-2])
print(s[::-1])
```

```python
# float div
print(type(6 / 2))

# floor div
print(type(6 // 2))

# Bitwise operator
a = 10
b = 4
print(a & b) 
print(a | b) 
print(~a) 
print(a ^ b) 
print(a >> 2) 
print(a << 2) 

print("------------------------------------------")
# Assignment operator
a = 10
b = a 
print(b) 
b += a 
print(b) 
b -= a 
print(b) 
b *= a 
print(b) 
b <<= a 
print(b)

print("-----------------------------------------")
#Identity operators
a = 10
b = 20
c = a 

print(a is not b) 
print(a is c)


print("------------------------------------------")
# Membership Operators 

x=10
y=34
list = [10,20,30,40]

print(x in list) # True
print(y not in list) # True


print("---------------------------------------------")
# Identity(is, is not) vs Equality(==)

# Equality operator
a=10
b=10

# Case 1:
# Return True because both a and b have the same value
print(a==b) 

# Case 2:
# Return True because both a and b is pointing to the same object   
print("Id of a: ",id(a))

print("Id of b: ",id(b))    

print(a is b)

# Case 3:
# Here variable a is assigned to new variable c, 
# which holds same object and same memory location
c=a       
print("Id of c: ",id(c))

print(a is c)
```
