# Hangman Game

```python
#  STEP-1

# TODO 1: Randomly choose a word from word_list and assign it to a variable known as chosen_word
import random

#Step 4

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']


word_list = ["Money", "Shine", "Active", "Crime", "Symphony", "Catstone"]
chosen_word = random.choice(word_list)
# print(chosen_word)


#TODO-1: - Create a variable called 'lives' to keep track of the number of lives left. 
#Set 'lives' to equal 6.
lives = 7


size = len(chosen_word)
    # print(size)

display = ["_"] * size
# print(display)

# TODO 2: Ask the user to provide input by guessing a letter and assign their answer to
# a varible guess.Make guess lower case.

# STEP 3:

# TODO: Use a while to make the user to guess again.The loop should only stop if the use guessed
# all the letters in chosen_word and display has no blank.Then you can tell user you won.

isWinner=False
while not isWinner:
    guess = input("Guess a letter: ").lower()
    # print(guess)

    # TODO 3: Check if the letter guessed by the user is one of the letter in chosen_word
    # for letter in chosen_word:
    #     if guess in letter:
    #         print("Right")
    #     else:
    #         print("Wrong")

    # STEP-2

    # TODO 1:Create an empty list called display.For each letter in the chosen_word add a "_" to
    # display.
    # TODO 2:Loop through each position of the chosen_word,if the letter at that position matches 
    # "guess" then reveal that letter at the display at that position.

    for idx in range(len(chosen_word)):
        #print(f"current pos:{idx}, current letter:{chosen_word[idx]}, guessed: {guess}")
        if chosen_word[idx].lower() == guess:
            display[idx] = guess
            
    #TODO-2: - If guess is not a letter in the chosen_word,
    #Then reduce 'lives' by 1. 
    if guess not in chosen_word:
        lives=lives-1
        #If lives goes down to 0 then the game should stop and it should print "You lose."
        #Join all the elements in the list and turn it into a String.
        if lives==0:
            print("You lose")
            break
        
            
    # TODO 3: Print the display and you should see the guess letter at the given a=index and rest
    # of index are "_"
    print(display)
    
    
    #The loop should only stop if the use guessed
    # all the letters in chosen_word and display has no blank.Then you can tell user you won.
    if "_" not in display:
        print("You win!!")
        isWinner = True

    #TODO-3: - print the ASCII art from 'stages' that corresponds
    #to the current number of 'lives' the user has remaining.
    print(stages[lives])

```
